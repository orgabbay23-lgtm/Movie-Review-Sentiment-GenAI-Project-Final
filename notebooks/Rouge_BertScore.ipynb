# ROUGE and BERTScore Evaluation

This notebook provides an interactive tool to evaluate the quality of generated review text against reference reviews using two complementary metrics:

- **ROUGE** (ROUGE-1, ROUGE-L) -- Measures n-gram overlap between the generated summary and the reference text. Higher scores indicate greater lexical similarity.
- **BERTScore** -- Measures semantic similarity using contextual embeddings. Captures meaning even when different words are used.

These metrics complement the Judge model's sentiment evaluation: a generated review can score high on ROUGE/BERTScore (semantically similar) yet still miss the target sentiment intensity.

```python
# Install necessary libraries
!pip install rouge-score
!pip install bert-score

from rouge_score import rouge_scorer
from bert_score import score

# Initialize the scorer once outside the loop
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

while True:
    print("\n--- חדש: הכנס טקסט מקורי וסיכום שנוצר ---")
    generated_summary = input("הכנס את הסיכום שנוצר (או 'exit' ליציאה): ")
    if generated_summary.lower() == 'exit':
        break

    reference_text = input("הכנס את הטקסט המקורי (או 'exit' ליציאה): ")
    if reference_text.lower() == 'exit':
        break

    # --- חלק 1: חישוב ROUGE (בדיקת חפיפת מילים) ---
    scores = scorer.score(reference_text, generated_summary)

    print("\n--- ROUGE Scores ---")
    print(f"ROUGE-1 (F1): {scores['rouge1'].fmeasure:.4f}") # חפיפת מילים
    print(f"ROUGE-L (F1): {scores['rougeL'].fmeasure:.4f}") # מבנה משפט

    # --- חלק 2: חישוב BERTScore (בדיקת משמעות) ---
    try:
        # BERTScore might take some time to download the model for the first run
        P, R, F1 = score([generated_summary], [reference_text], lang="en", verbose=False) # verbose=False to reduce output noise
        print("\n--- BERTScore ---")
        print(f"F1 Score: {F1.mean().item():.4f}")
    except Exception as e:
        print(f"An error occurred during BERTScore calculation: {e}")
        print("Please ensure the texts are in English or specify the correct language for BERTScore.")

print("יציאה מהלולאה. תודה!")
```
