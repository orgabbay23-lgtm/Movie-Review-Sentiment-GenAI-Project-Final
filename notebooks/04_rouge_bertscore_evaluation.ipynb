# ROUGE and BERTScore Evaluation

Use this notebook for manual spot checks between a reference review block and a generated representative review. It complements the learned judge model with lexical-overlap and semantic-similarity metrics.


```python
from bert_score import score
from rouge_score import rouge_scorer

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

while True:
    print('\nEnter a generated summary and its reference text to compute ROUGE and BERTScore.')
    generated_summary = input("Generated summary (or 'exit' to stop): ")
    if generated_summary.lower() == 'exit':
        break

    reference_text = input("Reference text (or 'exit' to stop): ")
    if reference_text.lower() == 'exit':
        break

    scores = scorer.score(reference_text, generated_summary)
    print('\n--- ROUGE Scores ---')
    print(f"ROUGE-1 (F1): {scores['rouge1'].fmeasure:.4f}")
    print(f"ROUGE-L (F1): {scores['rougeL'].fmeasure:.4f}")

    try:
        _, _, f1 = score([generated_summary], [reference_text], lang='en', verbose=False)
        print('\n--- BERTScore ---')
        print(f'F1 Score: {f1.mean().item():.4f}')
    except Exception as exc:
        print(f'An error occurred during BERTScore calculation: {exc}')
        print('Please ensure the texts are in English or specify the correct language for BERTScore.')

print('Exiting evaluation loop.')
```
